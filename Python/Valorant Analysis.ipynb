```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
vctdata = pd.read_csv("valorant champions istanbul.csv")
vctdata.head()
```

```python
vctdata.columns
```

```python
vctdata.shape
```

```python
vctdata['Prize'] = vctdata["Prize"].str.replace("$", "")
vctdata['Prize'] = vctdata["Prize"].str.replace(",", "")
vctdata.Prize = pd.to_numeric(vctdata.Prize)
```

```python
vctdata.boxplot(column=['Kill'])
```

```python
vctdata.boxplot(column=['Rounds Played'])
```

```python
vctdata.isnull().sum() # No missing data
```

```python
vctdata.describe
```

```python
vctdata['Nationality'] = vctdata['Nationality'].replace('South Corea', 'South Korea')
vctdata['Nationality'] = vctdata['Nationality'].replace('Sweeden', 'Sweden')
vctdata['Nationality']
```

```python
vctdata.describe()
```

```python
### We can expect to see some sort of skewness with the distributions of the variables. 
```

```python
vctdata['Nationality'].value_counts().plot(kind='bar', color='Green')
plt.title('Frequency of Nationalities of VCT Participants')
plt.xlabel('Nationalities')
plt.ylabel('Number of Individuals')
plt.show()
```

```python
plt.figure(figsize=(20,10))
sns.lmplot(x='K/D', y='Prize', data=vctdata, fit_reg=True)
plt.title("Relation between K/D and Prize", fontsize=16)
plt.xlabel("K/D")
plt.ylabel("Prize")
plt.show()
```

```python
plt.figure(figsize=(30,10))
sns.lmplot(x='HS %', y='Prize', data=vctdata, fit_reg=True)
plt.title("Relation between K/D and Prize", fontsize=16)
plt.xlabel("HS %")
plt.ylabel("Prize")
plt.show()
```

```python
plt.figure(figsize=(50,40))
sns.barplot(x = 'Player', y = 'Prize', data = vctdata)
plt.title("Prize won by VCT Participants", fontsize=48)
plt.xlabel("Player",fontsize=24)
plt.ylabel("Prize",fontsize=24)
```

```python
plt.figure(figsize=(50,20))

sns.scatterplot(x='Player', y='Kill', data=vctdata)

plt.title("Relation between Player and Kils", fontsize=48)
plt.xlabel("Player", fontsize=24)
plt.ylabel("Kill", fontsize=24)
plt.show()
```

```python
plt.figure(figsize=(20,10))

sns.scatterplot(x='Team', y='Rounds Win', data=vctdata)

plt.title("Relation between Team and Rounds Win", fontsize=48)
plt.xlabel("Player", fontsize=24)
plt.ylabel("Rounds Won", fontsize=24)
plt.show()
```
